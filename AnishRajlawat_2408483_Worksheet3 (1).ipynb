**Which passenger had the highest fare paid relative to their age?**

1. Add a column to the Titanic dataset, fare per year, containing the fare divided by the age of
the passenger(i.e., Fare/Age).</br>
2. Subset rows where fare per year is higher than 5, assigning this to high fare age.</br>
3. Sort high fare age by descending fare per year, assigning this to high fare age srt.</br>
4. Select only the Name and fare per year columns of high fare age srt and save the result as result.</br>
5. Look at the result.

```python
import pandas as pd

#load the titanic dataset
titanic_data = pd.read_csv('/content/drive/MyDrive/Datasets/Titanic-Dataset.csv')

#calculate fare per year and add it as a new column
titanic_data['Fare_per_Year'] = titanic_data['Fare'] / titanic_data['Age']

#filter for passengers with fare per year higher than 5
high_fare_age = titanic_data[titanic_data['Fare_per_Year'] > 5]

#sort by descending fare per year
high_fare_age_srt = high_fare_age.sort_values(by='Fare_per_Year', ascending=False)

#select the 'Name' and 'Fare_per_Year' columns
result = high_fare_age_srt[['Name', 'Fare_per_Year']]

#print the top 5 results
result.head()
```

**Which adult male passenger (age ≥ 18 and Sex is ’male’) paid the highest fare relative to their
class?</br>
To answer the question perform following operations:**

1. Add a column to the Titanic dataset, fare per class, containing the fare divided by the passenger
class i.e. Fare / Pclass.
2. Subset rows where the passenger is male (Sex is ”male”) and an adult (Age is greater than or equal

3. Sort adult males by descending fare per class, assigning this to adult males srt.
4. Select only the Name, Age, and fare per class columns of adult males sr and save the result

```python
import pandas as pd

#load the titanic dataset
titanic_data = pd.read_csv('/content/drive/MyDrive/Datasets/Titanic-Dataset.csv')

#calculate fare per class and add it as a new column
titanic_data['Fare_per_Class'] = titanic_data['Fare'] / titanic_data['Pclass']

#filter for adult male passengers
adult_males = titanic_data[(titanic_data['Sex'] == 'male') & (titanic_data['Age'] >= 18)]

#sort adult males by descending fare per class
adult_males_srt = adult_males.sort_values(by='Fare_per_Class', ascending=False)

#select the 'Name', 'Age', and 'Fare_per_Class' columns
result = adult_males_srt[['Name', 'Age', 'Fare_per_Class']]

#print the top 5 results
result.head()
```

**3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:</br>
Based on the dataset Answer the following question:</br>
What percent of the total fare revenue came from each passenger class?</br>
To answer the question perform following operation:**

```python
import pandas as pd

#load the titanic dataset
titanic_data = pd.read_csv('/content/drive/MyDrive/Datasets/Titanic-Dataset.csv')

#1. calculate total fare paid across all passengers
total_fare = titanic_data['Fare'].sum()
print("Total Fare:", total_fare)

#2. calculate total fare for first class passengers
first_class_fare = titanic_data[titanic_data['Pclass'] == 1]['Fare'].sum()
print("First Class Fare:", first_class_fare)

#3. calculate total fare for second class passengers
second_class_fare = titanic_data[titanic_data['Pclass'] == 2]['Fare'].sum()
print("Second Class Fare:", second_class_fare)

#4. calculate total fare for third class passengers
third_class_fare = titanic_data[titanic_data['Pclass'] == 3]['Fare'].sum()
print("Third Class Fare:", third_class_fare)

#5. combine fare totals into a list
class_fares = [first_class_fare, second_class_fare, third_class_fare]

#6. calculate the proportion of fare revenue by class
fare_proportions = [fare / total_fare for fare in class_fares]
print("Fare Proportions by Class:", fare_proportions)
```

**What percent of the total number of passengers on the Titanic belonged to each age group
(e.g., child, adult, senior)?</br>
To answer the question perform following operation:**

1. Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age
18{64), and "senior" (age 65 and above).</br>
2. Calculate the total number of passengers on the Titanic.</br>
3. Count the number of passengers in each age group.</br>
4. Divide the count of each age group by the total number of passengers to get the proportion of passengers
in each age group.</br>
5. Display the proportion as a percentage.

```python
import pandas as pd

#load the titanic dataset
titanic_data = pd.read_csv('/content/drive/MyDrive/Datasets/Titanic-Dataset.csv')

#1. create age group column
def categorize_age(age):
    if age < 18:
        return 'child'
    elif age < 65:
        return 'adult'
    else:
        return 'senior'

titanic_data['Age_Group'] = titanic_data['Age'].apply(categorize_age)

#2. calculate total number of passengers
total_passengers = len(titanic_data)

#3. count passengers in each age group
child_count = titanic_data['Age_Group'].value_counts()['child']
adult_count = titanic_data['Age_Group'].value_counts()['adult']
senior_count = titanic_data['Age_Group'].value_counts()['senior']

#4. calculate proportions
child_proportion = child_count / total_passengers
adult_proportion = adult_count / total_passengers
senior_proportion = senior_count / total_passengers

#5. display proportions as percentages
print("Proportion of children:", child_proportion * 100, "%")
print("Proportion of adults:", adult_proportion * 100, "%")
print("Proportion of seniors:", senior_proportion * 100, "%")
```
